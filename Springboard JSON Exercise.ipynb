Next, we will explore the data by flatten the nested element in the 'mjtheme_namecode' column.

```python
#Loads the file as text.
df = json.load((open('data/world_bank_projects.json')))

#Flatten the nested data in the major project theme column.
df_norm = json_normalize(df, 'mjtheme_namecode')

#Print the first 20 rows.
df_norm.head(20)
```

As we can see, some of the rows have missing values in the 'name' column. Instead of grouping the major project themes by names, we can group by the theme codes below. 

```python
#Count the rows by code and sort.
df_norm.groupby('code').count().sort_values('name', ascending = False)
```

By grouping the data by code, we find no missing values in that column. But we would still like to group and see the counts by 

```python
#Sort the rows by both code and missing values.
df_norm = df_norm.sort_values(['code','name'], ascending = True)

#Fill missing cells in name column with NaNs.
df_norm[df_norm['name'] == ''] = np.nan

#Backward fill the NaNs with existing code names.
df_norm = df_norm.fillna(method='bfill')

#Print the result
df_norm
```

```python
#Sort the count again by the theme names.
df_theme_count= df_norm.groupby('name').count().sort_values('code',ascending = False)

#Print the result again.
df_theme_count
```

The top 10 major project themes are:
+ Environment and natural resources management	
+ Rural development
+ Human development
+ Public sector governance
+ Social protection and risk management
+ Financial and private sector development
+ Social dev/gender/inclusion
+ Trade and integration
+ Urban development
+ Economic management
